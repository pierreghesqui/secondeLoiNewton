## Etape 2 - Ecrire les mesures : 
On lâche une balle et on mesure les positions successives de cette dernière grâce à une caméra : 

Commençons par écrire les valeurs de positions de la balle au cours du temps

✍️ **Complète le code ci-dessous en respectant les instructions**

```python
import numpy as np #NE PAS MODIFIER CETTE LIGNE. Cette ligne importe les extensions python pour faire des calculs
import matplotlib.pyplot as plt #NE PAS MODIFIER CETTE LIGNE. Cette ligne importe les extensions python 
                                #pour faire des graphiques
z = np.array([0, 0.002, 0.008, 0.018, 0.031, 0.049, 0.071, 0.096, 0.126,0.159, 0.196   ]) 
          #Ecris les valeur des positions successives de la balle 
          #Assure-toi que les valeurs sont bien écrites en mètre (et non pas en cm)

deltat =0.020  #Complète cette ligne en indiquant la durée (en seconde) séparant deux positions successives de la balle   
nbMesures = np.size(z) #NE PAS MODIFIER CETTE LIGNE. Elle calcule le nombre de mesures que tu as écrites dans z
t = [i*deltat for i in range(0,nbMesures)] #NE PAS MODIFIER CETTE LIGNE. 

print("Liste des positions : z =  ", z) #NE PAS MODIFIER CETTE LIGNE. Cette ligne affiche la liste des positions
print("La durée séparant 2 positions successives vaut: deltat =  ", deltat, " s") #NE PAS MODIFIER CETTE LIGNE.
print("Le nombre de mesures vaut : ", nbMesures) #NE PAS MODIFIER CETTE LIGNE
print("La liste des temps (en seconde): t = ",t)
```

✍️ **Exécute le code ci-dessus en cliquant dessus puis sur la flèche en haut de la page :**

## Etape 3 - Calculer les vitesses : 
✍️ **Complète le code ci-dessous pour calculer l'ensemble des vitesses en chaque point**

```python
vitesses = [] #NE PAS MODIFIER CETTE LIGNE. Création d'une liste vide pour stocker les valeurs de vitesses
for i in range(0,nbMesures-1): #NE PAS MODIFIER CETTE LIGNE. Cette ligne sert à répéter les calculs de vitesse v1, v2, v3...
    vitesse_i = (z[i+1]-z[i])/deltat   #Modifie cette ligne pour que python calcule la vitesse entre les points Mi et Mi+1
    vitesses.append(vitesse_i) #NE PAS MODIFIER CETTE LIGNE. Elle ajoute la vitesse calculée précédemment à la liste des vitesses

print("Les valeurs des vitesses (en m/s) sont : ", vitesses)  #NE PAS MODIFIER CETTE LIGNE. Elle affiche la valeur des vitesses
```

## Etape 3 - Calculer les accélérations :

✍️ **Complète le code ci-dessous pour calculer les valeurs des accélérations $\frac{\Delta v}{\Delta t}$**

```python
accelerations = [] #NE PAS MODIFIER CETTE LIGNE. Création d'une liste vide pour stocker les valeurs des accélérations
for i in range(0,nbMesures-2): #NE PAS MODIFIER CETTE LIGNE. Elle permet de répéter les mesures pour tous les points
    acceleration_i = (vitesses[i+1]-vitesses[i])/deltat   #Modifie cette ligne pour que python calcule la vitesse entre les points Mi et Mi+1
    accelerations.append(acceleration_i) #NE PAS MODIFIER CETTE LIGNE. Elle ajoute la vitesse calculée précédemment à la liste des vitesses

print("Les valeurs des accélérations (en mxs^-2) sont : ", accelerations)#NE PAS MODIFIER CETTE LIGNE
```

On estime que l'incertitude des mesures précédentes est d'environ $\pm 3 m\cdot s^{-2}$ <br>

## Etape 4 - Réduction de l'incertitude des mesures :
Pour réduire les incertitudes de mesures, on peut prendre la moyenne de l'ensemble des mesures de g calculées précédemment. 
L'incertitude vaudra alors : $\text{incertitude} = \frac{\text{écart-type}}{\text{nombre de mesures} -1}$<br>
Pour calculer la moyenne des accélérations, on peut utiliser la fonction <code>np.mean(acceleration)</code><br>
Pour calculer l'écart-type des accélérations, on peut utiliser la fonction <code>np.std(acceleration)</code>

✍️ **Complète le code ci-dessous** :

```python
moyenne = np.mean(accelerations) #Complète cette ligne
incertitude = np.std(accelerations)/(nbMesures-1)**0.5 #Complète cette ligne
print("g = ", moyenne, " +/- ", incertitude) #NE PAS MODIFIER CETTE LIGNE
```
